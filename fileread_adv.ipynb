```python
import csv

file = open("students.csv","r")
reader = csv.DictReader(file)

print("reading student data.....\n")

for row in reader:
    print(f"Name: {row['name']}, Age: {row['age']}, Grade: {row['grade']}")

file.close()
```

```python
# writing to the csv file
import csv

new_students = [{"id":5,"name":"John", "age":25, "grade":"C"},
                {"id":6,"name":"Alice", "age":18, "grade":"A"}]

file = open("students.csv","w", newline="")
writer = csv.DictWriter(file, fieldnames=["id","name","age","grade"])
# writer.writeheader()
writer.writerows(new_students)
file.close()

print("new student records saved")
```
