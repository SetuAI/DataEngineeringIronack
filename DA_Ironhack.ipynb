```python
print("get the missing values.....\n")

df.isnull().sum()
```

```python

# alternate method to get columns with missing data

columns_with_missing = df.columns[df.isnull().any()].tolist()

print(columns_with_missing)
```

```python
# we want to find in percentage - missing values
missing_percentage = (df.isnull().sum()/len(df)) * 100

print(round(missing_percentage,2))
```

```python
total_missing_values_df = df.isnull().sum().sum()

print(f"Entire dataframe has {total_missing_values_df} missing values")
```
